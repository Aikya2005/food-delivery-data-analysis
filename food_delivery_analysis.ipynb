```python
import pandas as pd
import numpy as np
orders = pd.DataFrame({
    "order_id": range(1, 1001),
    "user_id": np.random.randint(1, 301, 1000),
    "restaurant_id": np.random.randint(1, 51, 1000),
    "total_amount": np.random.randint(200, 800, 1000),
    "order_date": pd.date_range(start="2023-01-01", periods=1000, freq="D")
})
users = pd.DataFrame({
    "user_id": range(1, 301),
    "city": np.random.choice(["Bangalore", "Hyderabad", "Chennai", "Pune"], 300),
    "membership": np.random.choice(["Gold", "Regular"], 300)
})
restaurants = pd.DataFrame({
    "restaurant_id": range(1, 51),
    "restaurant_name": [f"Restaurant {i}" for i in range(1, 51)],
    "cuisine": np.random.choice(["Indian", "Chinese", "Italian", "Mexican"], 50),
    "rating": np.round(np.random.uniform(3.0, 5.0, 50), 1)
})

orders.head(), users.head(), restaurants.head()
df = orders.merge(users, on="user_id", how="left")
df = df.merge(restaurants, on="restaurant_id", how="left")

df.head()
gold_orders = df[df["membership"] == "Gold"]
gold_orders.shape[0]
df.groupby("city")["total_amount"].sum().sort_values(ascending=False)


```
